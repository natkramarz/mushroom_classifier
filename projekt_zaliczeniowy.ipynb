- veil-type (rodzaj pochwy grzyba): partial=p (częściowa), universal=u (uniwersalna)

- veil-color (kolor pochwy grzyba): brown=n (brązowy), orange=o (pomarańczowy), white=w (biały), yellow=y (żółty)

- ring-number (liczba pierścieni grzyba): none=n (brak), one=o (jeden), two=t (dwa)

- ring-type (rodzaj pierścienia): cobwebby=c (pajęczynowy), evanescent=e (zanikający), flaring=f (kloszowy),large=l (duży), none=n (brak), pendant=p (wiszący), sheathing=s ( osłaniający), zone=z (sfera)

- spore-print-color (kolor zarodników grzyba): black=k (czarny), brown=n (brązowy), buff=b (płowy), chocolate=h (brązowy), green=r (zielony), orange=o (pomarańczowy), purple=u (fioletowy), white=w (biały), yellow=y (żółty)

- population (populacja): abundant=a (przepełniona), clustered=c (zgrupowana), numerous=n (liczna),scattered=s (rozsiana), several=v (kilka), solitary=y (samotny)

## Przetworzenie danych

Usuwam kolumny z informacją o populacji grzyba, ponieważ grzybiarz oceniający cechy grzyba może nie mieć dostępu do takich informacji 

```python
df.drop("population", inplace=True, axis=1)

df.head()
```

Usuwam rekordy z brakującymi informacjami o atrybutach

```python
import numpy as np
df = df.replace('', np.nan)   
df.dropna()
```

Usuwam duplikaty:

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
pd.set_option('display.max_columns', None)
df.describe()
```

Jak możemy zauważyć w kategorii rodzaju koloru pochwy grzyba pojawia się tylko jedna kategoria, czyli ta kolumna nie przyda nam się w dalszych obliczeniach. Usunę ją. 

```python
df.drop(["veil-type"], axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

df_col = df["class"].replace({'e': 'edible', 'p': 'poisonous'})
plt.hist(df_col.values, bins=2, color='skyblue', ec='k')
```

Zmieniam kategorie na wartości numeryczne

```python
for column in df.columns:
    df[column] = pd.Categorical(df[column])
    print(f'column: {column}, mappings: {dict( enumerate(df[column].cat.categories ) )}')
    df[column] = df[column].cat.codes
df.head()
```

```python
import seaborn as sns 

fig, ax = plt.subplots(figsize=(20, 20))

sns.heatmap(df.corr(), cmap="RdBu", vmin=-1, vmax=1, annot=True, center=0, square=True, linewidth=4, ax=ax)
plt.show()
```

Widzimy, że największa korelacja dodatnia zmiennej wyjściowej istnieje ze zmienną gill-color (rozmiar blaszki). Faktycznie w większości przypadków szerokie blaszki wskazują na grzyby jadalne, a z kolei wąskie blaszki wskazują na grzyby trujące. 

```python
df_show = df.copy()
df_show["gill-size"].replace({0: 'broad', 1: 'narrow'}, inplace=True) 
df_show["class"].replace({0: 'edible', 1: 'poisonous'}, inplace=True) 
sns.countplot(data = df_show, x = 'class', hue = 'gill-size')
plt.show()
df_show
```

Również duża korelacja występuje pomiędzy zmienną gill-color (kolor blaszki), a zmienną wyjściową. Z wykresu na dole możemy odczytać, że 60% grzybów trujących ma blaszki koloru płowego.  

```python
df_show["gill-color"].replace({0: 'buff', 1: 'red', 2: 'grey', 3: 'chocolate', 4: 'black', 5: 'borwn', 6: 'orange', 7: 'pink', 8: 'green', 9: 'purple', 10: 'white', 11: 'yellow'}, inplace=True) 
sns.countplot(data = df_show, x = 'class', hue = 'gill-color')
plt.show()
```

Z kolei pomiędzy zmienną veil-color, a zmienną gill-attachment jest korelacja 0.85, obie korelacje ze zmienną class są małe, stąd usuwam jedną z nich o mniejszej korelacji ze zmienną class - gill-attachment

```python
df_show["veil-color"].replace({0: 'brown', 1: 'orange', 2: 'white', 3: 'yellow'}, inplace=True) 
df_show["gill-attachment"].replace({0: 'attached', 1: 'free'}, inplace=True)
sns.countplot(data = df_show, x = 'veil-color', hue = 'gill-attachment')
plt.show()

df.drop(["gill-attachment"], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['class'], axis=1)
y = df['class']


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)
```

## Modelowanie

```python
models = []
model_names = ['Logistic Regression', 'SVM Linear', 'SVM Poly', 'SVM rbf', 'Decision Tree Classifier', 'Random Forest Classsifier', 'MLP Classifier', 'Keras Neural Network']
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

pipe1 = Pipeline([('preprocessing', StandardScaler()), ('classifier', LogisticRegression(max_iter=3000, class_weight='balanced'))])

param_grid1 = {
            'preprocessing': [StandardScaler(), None],
            'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100, 1000], 
}

from sklearn.model_selection import StratifiedKFold

seed=123
kfold = StratifiedKFold(n_splits=5, random_state=seed, shuffle=True)

logistic_regression = GridSearchCV(pipe1, param_grid1, cv=kfold, return_train_score=True)

logistic_regression.fit(X_train, y_train)
models.append(logistic_regression.best_estimator_)
logistic_regression.best_params_
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

pipe2 = Pipeline([('preprocessing', StandardScaler()), ('classifier', SVC(kernel='linear', class_weight='balanced'))])

param_grid2 = {
            'preprocessing': [StandardScaler(), None],
            'classifier__gamma': [0.001, 0.01, 0.1, 1, 10, 100],
            'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100]
}

svc_linear = GridSearchCV(pipe2, param_grid2, cv=kfold, return_train_score=True)

svc_linear.fit(X_train, y_train)
models.append(svc_linear.best_estimator_)
svc_linear.best_params_
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

pipe3 = Pipeline([('preprocessing', StandardScaler()), ('classifier', SVC(kernel="poly", class_weight='balanced'))])

param_grid3 = {
            'preprocessing': [StandardScaler(), None],
            'classifier__degree': [1, 2, 3, 3],
            'classifier__coef0': [0], 
            'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100]
}

svc_poly = GridSearchCV(pipe3, param_grid3, cv=kfold, return_train_score=True)

svc_poly.fit(X_train, y_train)
models.append(svc_poly.best_estimator_)
svc_poly.best_params_
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC

pipe4 = Pipeline([('preprocessing', StandardScaler()), ('classifier', SVC(kernel="rbf", class_weight='balanced'))])

param_grid4 = {
            'preprocessing': [StandardScaler(), None],
            'classifier__gamma': [0.001, 0.01, 0.1, 1, 10, 100],
            'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100]
}

svc_rbf = GridSearchCV(pipe4, param_grid4, cv=kfold, return_train_score=True)

svc_rbf.fit(X_train, y_train)
models.append(svc_rbf.best_estimator_)
svc_rbf.best_params_
```

```python
from sklearn import tree

pipe5 = Pipeline([('preprocessing', StandardScaler()), ('classifier', tree.DecisionTreeClassifier(class_weight='balanced'))])

param_grid5 = {
            'preprocessing': [StandardScaler(), None],
            'classifier__max_depth': [3, 5, 10, 15, 20],
            'classifier__max_leaf_nodes': [10, 20, 30, 40, 50],
            'classifier__criterion': ['gini', 'entropy'],
}

decision_tree = GridSearchCV(pipe5, param_grid5, cv=kfold, return_train_score=True)

decision_tree.fit(X_train, y_train)
models.append(decision_tree.best_estimator_)
decision_tree.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier

pipe6 = Pipeline([('preprocessing', StandardScaler()), ('classifier', RandomForestClassifier(class_weight='balanced'))])

param_grid6 = {
            'preprocessing': [StandardScaler(), None],
            'classifier__n_estimators': [3, 4, 6, 16, 32, 64],
            'classifier__max_leaf_nodes': [10, 20, 30, 40, 50],
}

random_forest = GridSearchCV(pipe6, param_grid6, cv=kfold, return_train_score=True)

random_forest.fit(X_train, y_train)
models.append(random_forest.best_estimator_)
random_forest.best_params_
```

```python
from sklearn.neural_network import MLPClassifier

pipe7 = Pipeline([('preprocessing', StandardScaler()), ('classifier', MLPClassifier())])

param_grid7 = {
            'preprocessing': [StandardScaler(), None],
            'classifier__hidden_layer_sizes': [(20,10)],
            'classifier__learning_rate_init': [0.001, 0.01, 0.1],
            'classifier__max_iter': [1000, 2000],
            'classifier__batch_size': [8, 16,32],
}

mlp_clf = GridSearchCV(pipe7, param_grid7, cv=kfold, return_train_score=True)

mlp_clf.fit(X_train, y_train)
models.append(mlp_clf.best_estimator_)
mlp_clf.best_params_
```

```python
from keras.models import Sequential
from keras.layers import Dense

from keras.callbacks import History

history = History()
keras_model = Sequential()
keras_model.add(Dense(100,activation="sigmoid", input_shape=(X_train.shape[1],)))
keras_model.add(Dense(50,activation="sigmoid"))
keras_model.add(Dense(10,activation="sigmoid"))
keras_model.add(Dense(1,activation="sigmoid"))
keras_model.summary()

keras_model.compile(loss="binary_crossentropy", optimizer="Adam", metrics=["accuracy"])
X_train_keras, X_keras_val, y_train_keras, y_keras_val = train_test_split(X_train, y_train, test_size=0.4, random_state=0)

history = keras_model.fit(X_train_keras, y_train_keras, validation_data=(X_keras_val, y_keras_val), batch_size=32, epochs=100)

models.append(keras_model)
```

## Wizualizacja wyników, porównanie modeli

```python
from sklearn.metrics import confusion_matrix
import tensorflow as tf

fig, axs = plt.subplots(2, 4, figsize=(20,10))

def get_confusion_matrix(model, model_index):
    
    if index == 7:
        cm = confusion_matrix(y_test, [round(x[0]) for x in model.predict(X_test)])
    else:
        cm = confusion_matrix(y_test, model.predict(X_test))
    i = 0
    j = model_index
    if model_index > 3:
        i = 1
        j -= 4

    sns.heatmap(cm, annot=True, fmt='g', ax=axs[i, j], cmap="Greens")
    axs[i, j].set_xlabel('Predicted labels')
    axs[i, j].set_ylabel('True labels') 
    axs[i, j].set_title(model_names[model_index])
    axs[i, j].xaxis.set_ticklabels(['edible', 'poisonous'])
    axs[i, j].yaxis.set_ticklabels(['edible', 'poisonous'])


for index, model in enumerate(models):
    get_confusion_matrix(model, index)
plt.show()
```

```python
from sklearn import  metrics

precision_score = []
recall_score = []
f1_score = []
accuracy_score = []

for index, model in enumerate(models):
    if index == 7:
        predict_x=model.predict(X_test) 
        classes_x = np.argmax(predict_x,axis=1)
    else:
        prediction = model.predict(X_test)
    precision_score.append(metrics.precision_score(y_test, prediction, average='weighted'))
    recall_score.append(metrics.recall_score(y_test, prediction, average='weighted'))
    f1_score.append( metrics.f1_score(y_test, prediction, average='weighted'))
    accuracy_score.append(metrics.accuracy_score(y_test, prediction))



d = {'precision_score': precision_score, 
     'recall_score': recall_score, 
     'f1_score': f1_score,
     'accuracy_score' : accuracy_score
    }
df = pd.DataFrame(data=d)
df.insert(loc=0, column='Model', value=model_names)
df
```

Porównajmy feature importance dla Decision Tree Classifier i Random Forest Classifier

```python
print(decision_tree.score(X_test, y_test))

def plot_feature_importances(model, data):
    n_features = data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), data.columns)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.ylim(-1, n_features)

plt.figure(figsize=(10, 10))
plt.title("Decision Tree Classifier")
plot_feature_importances(decision_tree.best_estimator_.steps[1][1], X)
plt.show()

plt.figure(figsize=(10, 10))
plt.title("Random Forest Classifier")
plot_feature_importances(random_forest.best_estimator_.steps[1][1], X)
plt.show()
```
